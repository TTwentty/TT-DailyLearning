## python的模块、包与库
**模块与包**是有**严谨指向**的名词性概念；**库**是**宽泛说法**，强调其功能性。  
1. 模块（module）  
    是一个.py文件，文件中可能含有多个函数、变量、类等。
2. 包（package）  
    是一个**包含一个【\_\_init\_\_.py】文件** + 1或多个.py文件的文件夹。包目录下，可以有多个子包，文件夹目录中，**必须**要有初始化文件即\_\_init\_\_.py，才能使该文件夹成为一个包。  
   \_\_init\_\_.py中的短下划线是前两根、后两根。
4. 库  
    第三方提供，具有某种特定、完整的**功能**的**工具箱**。  
    可以只有一个模块（一个.py文件），但通常是一个包，且包含了许多子包，毕竟需要实现完整功能。

5. 使用：
    在其他.py文件中编写代码时，若要使用某库中的函数，在import操作后，调用函数大概是【库名.包名.函数名(输入参数)】这样调用的，其中库名其实就是第三方库的最大包，包名是指函数所处的具体子包的包名。

```python
def create_symmetric_matrix(n):
    """生成随机对称矩阵，对角线为0"""
    import numpy as np
    A = np.random.rand(n, n)
    A = (A + A.T)/2 # 对称化
    np.fill_diagonal(A, 0)
    return A
    
# 试试函数创建成功没
B = create_symmetric_matrix(3)
print(B)
```

列表是用的中括号

```python
raw_data = ["2306422", "1465.23", "N/A", "896.5233"]
clean_data = [float(x) if
             x.replace('.', ' ').isdigit()
             else None for x in raw_data]
clean_data = [x for x in clean_data if x is not None]
# 删除None
print(clean_data)
```

# Python中的文件读取
有多种方式可读取文件。

```python
# 以只读模式打开文件：
# 使用open()函数，提供两个参数，分别是文件名和模式，'r'代表只读模式
# <with>语句确保文件在使用后被正确关闭。

with open('555.txt', 'r') as file:
    content = file.read()
print(content)

# 文本文件【555.txt】如果是中文文字，输出是乱码，英文和数字可以输出；
# 只能打开工作目录的文件，就是与python文件一个文件夹的
# current working directory.
```

```python
# 逐行读取文件
# readline()或readlines(),前者每次读取文件的一行，后者一次读取所有行并返回一个列表

with open('555.txt', 'r') as file:
    for line in file:
        print(line.strip()) # strip()去除行尾的换行符
```
