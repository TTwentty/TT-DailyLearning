# 代码5.5 字符串索引和切片

```python
greeting_str = 'Hey, Freeanna!'
print('字符串长度为：')
print(len(greeting_str))

# 打印每个字符和对应的索引
# enumberate()函数返回一个迭代器，其中包含每个字符及其对应的索引
# for循环遍历上述迭代器，依次打印出每个字符和它们的索引
# print()括号中的f是在字符串中插入变量的用法，花括号中为所插入的变量
for index, char in enumerate(greeting_str):
    print(f"字符：{char}，索引：{index}")

# 单个字符索引，从0开始算，如果是负数，说明是从字符串末尾开始算
print(greeting_str[0])
print(greeting_str[1])

print(greeting_str[-1])
print(greeting_str[-2])

# 切片使用的语法为：字符串变量名[start:end:step]
# 其中end是不被包括的，相当于“左闭右开”

# 取出前3个字符，即索引为0,1,2的字符
print(greeting_str[:3])

# 取出索引为1,2,3,4,5的字符，即不包含0与6
print(greeting_str[1:6])

# 指定步长为2，取出的是0,2,4...一直到字符串的尽头
print(greeting_str[::2])

# 制定步长为-1，是从末尾开始取值
print(greeting_str[::-1])
```

# 代码5.6 将数据插入字符串
很多场合需要将数据插入字符串，比如：
- 使用print()时；
- 图片中插入图例（Legengd）、标题（title）、打印日期时间、打印统计量（均值、方差、标准差、四分位）等。

4种常用方法：
- +运算；
- % 占位符（类似C语言printf()使用的）；
- str.format()；
- f-strings，Python3.6版本开始引入的一种方式，f和F都可用。

```python
name = 'Freeanna'
height = 156.0

str_1 = name + ' has a height of ' + str(height) + ' cm.'
print(str_1)

str_2 = '%s has a height of %.3f cm.' %(name, height) # 单引号与占位索引之间可以留空格
print(str_2)

str_3 = '{} has a height of {:.3f} cm.' .format(name, height) # 单引号与.format()之间可以留空格
print(str_3)

str_4 = f'{name} has a height of {height:.3f} cm.' # f可以换成F
print(str_4)
```

# 代码5.7 列表索引和切片
列表的元素装在中括号（方括号，[]）中；<P>
列表中的元素用,隔开

```python
# 创建一个混合列表
my_list = [1, 1.0, '12ab', True, [1, 1.0, '1'], {1}, {1:1.0}]
print('列表长度为')
print(len(my_list))

# 打印每个元素和对应的索引
for index, item in enumerate(my_list):
    type_i = type(item)
    print(f"元素：{item}，  索引：{index}，  类型：{type_i}")

# 列表索引
print(my_list[0])
print(my_list[1])

print(my_list[-1])
print(my_list[-2])

# 列表切片
# 取出前3个元素，索引为0、1、2
print(my_list[:3])

# 取出索引1、2、3，不含0，不含4
print(my_list[1:4])

# 指定步长2，取出第0、2、4、6。可以用索引[::2]，因为总共有7个元素
print(my_list[0:7:2])

# 指定步长-1，倒序
print(my_list[::-1])

# 提取列表中的列表某个元素
print(my_list[4][1])
```

# 代码5.8 列表常用方法、操作

```python
# 创建一个混合列表
my_list = [1, 1.0, '12ab', True, [1, 1.0, '1'], {1}, {1:1.0}]
print(my_list)

# 修改某个元素
my_list[2] = '123'
print(my_list)

# 在列表指定位置插入元素
my_list.insert(2, 'inserted')
print(my_list)

# 在列表尾部插入元素
my_list.append('tail')
print(my_list)

# 通过索引删除
del my_list[-1]
print(my_list)

# 删除某个元素
my_list.remove('123')
print(my_list)

# 判断一个元素是否在列表中

if '123' in my_list:
    print('YES')
else:
    print("NO") # 用‘’和""是一样的

# 列表翻转
my_list.reverse()
print(my_list)

# 将列表中所有字符用下划线依次连接
letters = ['F', 'r', 'e', 'e', 'a', 'n', 'n', 'a']
word = '_'.join(letters)
print(word)
```
